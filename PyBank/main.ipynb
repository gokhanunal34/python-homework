```python
from pathlib import Path
import csv

#define the path for budget_data(I had to write down the whole path to eliminate the error I was getting)

budget_csv = Path("c:\MIA-VIRT-FIN-PT-03-2022-U-LOL/02-Python/Homework/02-Python/Instructions/PyBank/Resources/budget_data.csv")

# some variables and lists

line_num = 0
ploss = []
ploss_dates = []
ploss_sum = 0
max_pl = 0
min_pl = 0
sum = 0

#read the csv file 

with open(budget_csv, 'r') as csvfile:
    
   
    
    csv_reader = csv.reader(csvfile, delimiter = ',')
    
    
    
    header = next(csv_reader)
    
    
    #loop for going through the csv and appending the corresponding lists
    
    for row in csv_reader:
        
        line_num += 1
        ploss.append(int(row[1]))
        ploss_sum += int(row[1])
        ploss_dates.append(row[0])
        
#profit loss change part    
        
ploss_change = []

for i in range(1, len(ploss)):
    x = ploss[i] - ploss[i - 1]
    ploss_change.append(int(x))
    
#start range from 0 since ploss_change is a new list with no headers
for i in range(0, len(ploss_change)):
    sum += ploss_change[i]
    average_change_ploss = round((sum / (len(ploss_change))), 2) #the sample output seems like rounded 2 decimal places


#ploss now includes only numbers so we can call it without specifying in []
#this for loop first sets max_pl and min_pl to the first row of data and then goes through the list changing max and min everytime
   
for profloss in ploss:
    
    if min_pl == 0:
        max_pl == profloss
        min_pl == profloss
    if profloss > max_pl:
        max_pl = profloss
    elif profloss < min_pl:
        min_pl = profloss
        
#index max_pl and min_pl
max_pl_index = ploss.index(max_pl)
min_pl_index = ploss.index(min_pl)

#max_pl_date grabs whatever corresponds to max_pl_index in ploss_dates list
max_pl_date = ploss_dates[max_pl_index]
min_pl_date = ploss_dates[min_pl_index]

#output to the folder as .txt
output_path = Path('Financial_Analysis.txt')

with open(output_path, 'w') as file:
    file.write(f"Financial Analysis\n")
    file.write(f"----------------------------\n")
    file.write(f"Total Months: {line_num} \n")
    file.write(f"Total: ${ploss_sum}\n")
    file.write(f"Average Change: ${average_change_ploss}\n")
    file.write(f"Greatest Increase in Profits: {max_pl_date} (${max_pl}) \n")
    file.write(f"Greatest Decrease in Profits: {min_pl_date} (${min_pl}) \n")
```
